```python
def get_SU_norm_norm(a,b):
    #sigma_1=np.cov(a/np.sum(a))
    sigma_1=np.cov(a)
    sigma_1=np.expand_dims(sigma_1, axis=(0, 1))
    #sigma_2=np.cov(b/np.sum(b))
    sigma_2=np.cov(b)
    sigma_2=np.expand_dims(sigma_2, axis=(0, 1))
    #sigma_12=np.cov(a/np.sum(a),b/np.sum(b))
    sigma_12=np.cov(a,b)
    MI=np.log2(np.linalg.det(sigma_1)*np.linalg.det(sigma_2)/np.linalg.det(sigma_12))
    #H1=(np.log2(2*np.pi*sigma_1)+1).item()
    #H2=(np.log2(2*np.pi*sigma_2)+1).item()
    return MI/2
    
    
```

```python
def get_SU_norm_cls(a,t):#a is lianxi variable, t is the binaray target
    index_0=np.where(t==0)
    index_1=np.where(t==1)
    sigma_a=np.var(a)
    sigma_0=np.var(a[index_0])
    sigma_1=np.var(a[index_1])
    p_0=len(index_0[0])/len(t)
    p_1=1-p_0
    MI=(np.log2(2*np.pi*sigma_a)+1)-p_0*(np.log2(2*np.pi*sigma_0)+1)-\
         p_1*(np.log2(2*np.pi*sigma_1)+1)
    #H_cls=2*(-np.log2(p_0)*p_0-np.log2(p_1)*p_1)
    # p_0*(np.log2(2*np.pi*sigma_0)+1) 第0类的概率*第0类时候的条件熵
    #H_a=(np.log2(2*np.pi*sigma_a)+1)
    return MI/2
```

```python
def get_SU_norm_cls_norm(x,y1,y2):#x,y2 is lianxi variable, y1 is the binaray target
    sigma_x=np.var(x)
    sigma_y2=np.var(y2)
    index_0=np.where(y1==0)
    index_1=np.where(y1==1)
    p_0=len(index_0[0])/len(x)
    p_1=1-p_0
    Sigma_0=np.linalg.det(np.cov(x[index_0],y2[index_0]))
    Sigma_1=np.linalg.det(np.cov(x[index_1],y2[index_1]))
    MI=np.log2(sigma_x)+np.log2(sigma_y2)-p_0*np.log2(Sigma_0)-p_1*np.log2(Sigma_1)
    #H_x=(np.log2(2*np.pi*sigma_x)+1)
    #H_y1=2*(-np.log2(p_0)*p_0-np.log2(p_1)*p_1)
    #H_y2=(np.log2(2*np.pi*sigma_y2)+1)
    return MI/2#MI/2
    
    
```

```python
# def get_var_rate(a,range_list):# a is a feature and range_list is a range list
#     var=0
#     for range_index in range_list:
#         var+=np.var(a[range_index])
#     return 1-var/(np.var(a)*len(range_list))
def get_var_rate(a,range_list):
    mean = []
    for range_index in range_list:
        mean.append(np.mean(a[range_index]))
    return np.var(mean)/np.var(a)
```
